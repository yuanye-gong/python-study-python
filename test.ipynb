```python
import matplotlib.pyplot as plt
```

```python
dino_vectors = [(6,4), (3,1), (1,2), (-1,5), (-2,5), (-3,4), (-4,4), (-5,3), (-5,2), (-2,2),
                (-5,1), (-4,0), (-2,1), (-1,0), (0,-3), (-1,-4), (1,-4), (2,-3), (1,-2), (3,-1), (5,1)]
```

```python
for i in range(0,len(dino_vectors)):
    x1, y1 = dino_vectors[i]
    x2, y2 = dino_vectors[(i+1)%len(dino_vectors)]
    plt.scatter(*dino_vectors[i], color='C3')
    plt.plot([x1, x2], [y1, y2], color='C0')
```
